```python

class NeuralNetwork:
    def __init__(self):
        
        self.ranger=0
        self.eps=0.01961
        #parameters
        
        self.inputSize = 15
        self.hiddens_one_Size = 4
        self.hiddens_two_Size = 7
        self.outputSize = 1
        
        #weights
        self._weights_ = np.random.randn(self.inputSize, self.hiddens_one_Size) 
        self.__weights__ = np.random.randn(self.hiddens_one_Size, self.hiddens_two_Size)
        self.___weights___ = np.random.randn(self.hiddens_two_Size, self.outputSize) 
        

    def step_all(self, in_values):

        
        
        self.hiddens_one = np.dot(in_values, self._weights_) 
        self.hiddens_one_sigmoid = self.sigmoid(self.hiddens_one) 
        
        self.hiddens_two = np.dot(self.hiddens_one_sigmoid, self.__weights__) 
        self.hiddens_two_sigmoid = self.sigmoid(self.hiddens_two) 
        
        
        self.output = np.dot(self.hiddens_two_sigmoid, self.___weights___) 
        self.output_sigmoid = self.sigmoid(self.output)
        return self.output_sigmoid
        
    def sigmoid(self, values, deriv=False):
        return values * (1 - values) if deriv else np.exp(values)/(np.exp(values)+1)
    
    
    def mse(self, true, false):
        return np.mean(np.square(true - false)).values
      
    def weights_update(self,select_weights, config=True, learning_rate=0.0001, printer=False):
        
        if select_weights=='third':
            gradient=self.hiddens_two_sigmoid.T.dot(self.output_delta)
            if config:
                if printer:
                    print('output to hiddens_2')
                else:
                    pass
                return self.___weights___ +gradient*learning_rate
            else:
                return self.___weights___

        elif select_weights=='second':
            gradient=self.hiddens_one_sigmoid.T.dot(self.hiddens_two_delta)
            if config:
                if printer:
                    print('hiddents_2 to hiddens_1')
                else:
                    pass
                return self.__weights__ +gradient*learning_rate
            else:
                return self.__weights__
            
        elif select_weights=='first':
            gradient=in_values.T.dot(self.hiddens_one_delta)
            if config:
                if printer:
                    print('hiddents_1 to input \n')
                else:
                    pass
                return self._weights_ +gradient*learning_rate
            else:
                return self._weights_
            
            
            
        else:
            return 'error in select values'

    
    def backward(self, in_values, out_values, output_sigmoid,
                _config_=True,__config__=True,___config___=True):

        self.errors = out_values-output_sigmoid
        self.output_delta = self.errors * self.sigmoid(output_sigmoid, deriv=True)
        
        self.___weights___=self.weights_update(select_weights='third',config=___config___)
        
        
        self.hiddens_two_error = self.output_delta.dot(self.___weights___.T) 
        self.hiddens_two_delta = self.hiddens_two_error * self.sigmoid(self.hiddens_two_sigmoid, deriv=True)

        self.__weights__=self.weights_update(select_weights='second',config=__config__)
        
        
        self.hiddens_one_error = self.hiddens_two_delta.dot(self.__weights__.T) 
        self.hiddens_one_delta = self.hiddens_one_error * self.sigmoid(self.hiddens_one_sigmoid, deriv=True)

        
        self._weights_=self.weights_update(select_weights='first',config=_config_)
        

        
        return self._weights_ , self.__weights__, self.___weights___
        
        
    def train(self, in_values, out_values,
             _config_=True,__config__=True,___config___=True):
        
        output_sigmoid = self.step_all(in_values)
        
        self.backward(in_values, out_values,
                      output_sigmoid,
                      _config_=_config_,
                      __config__=__config__,
                      ___config___=___config___)
        
        
    def ReLu(self,predict):
        if predict>0.8:
            return 1
        elif predict<0.2:
            return 0
        else:
            return predict
        
```

```python
def main():
    NN = NeuralNetwork()
    ranger=0
    while True:
        ranger+=1
        if (ranger % 1000== 0):
            errors=NN.mse(true=out_values, false= NN.step_all(in_values))
            print('iteration--->'+str(ranger), "Loss: " , errors)
            if errors<NN.eps:
                
                break
        NN.train(in_values, out_values,
                ___config___=True,
                __config__=True,
                _config_=True)
    return NN
    
    
        
```
